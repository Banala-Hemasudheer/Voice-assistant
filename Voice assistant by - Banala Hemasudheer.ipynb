```python
# Import required libraries
import speech_recognition as sr
import pyttsx3
import datetime

# Duration set to 4 secs
duration=int(4)
print('Loading your Personal care taker - Danny')

# Choosing voice and it's rate
engine=pyttsx3.init()
voices=engine.getProperty('voices')
engine.setProperty('voice',voices[1].id)
engine.setProperty('rate',190) 

# Defining speak function
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Definig wishMe function
def wishMe():
    hour=datetime.datetime.now().hour
    if hour>=0 and hour<12:
        speak("Hello,Good Morning")
        print("Hello,Good Morning")
    elif hour>=12 and hour<16:
        speak("Hello,Good Afternoon")
        print("Hello,Good Afternoon")
    else:
        speak("Hello,Good Evening")
        print("Hello,Good Evening")

# Definig takeCommand function
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print('Please Speak')
        r.pause_threshold=1 # pausing time
        print('listening')
        audio=r.record(source,duration=duration)
        print('decoding')
        try:
            statement=r.recognize_google(audio,language='en-In')
            print(f"user said:{statement}\n")
        except Exception as e:
            speak("Pardon me, please say that again")
            return "None"
        return statement
```

```python
# Run the code below to talk with Danny
speak("Loading your Personal care taker Danny")
wishMe()

if __name__=='__main__':
        
    while True:                                         # while true means ifinite loop
        speak("Tell me how can I help you now?")
        statement = takeCommand().lower()
        if statement==0:
            continue

        if 'is your name' in statement in statement:
            speak('My name is Danny and I was programmed to act as personal care taker.')

        elif "good bye" in statement or "ok bye" in statement or "stop" in statement:
            speak('your personal care taker danny is shutting down,Good bye')
            print('your personal care taker danny is shutting down,Good bye')
            break

        elif 'who are you' in statement or 'what can you do' in statement:
            speak('I am Danny, your personal care taker. I am programmed to assist you'
            'You can ask me about current time and can tell me about your health condition to get suggestions')

        elif "feeling headache" in statement:
            speak("stop using mobile and take one ibuprofen tablet from cupboard, please maintain proper sleep pattern")

        elif "having fever" in statement:
            speak("you can find dolo 650 tablets in the cupboard, take one and have some rest")

        elif "having cold" in statement or 'running nose' in statement:
            speak("take one benadryl tablet from cupboard")    

        elif "who made you" in statement or "who created you" in statement or "who discovered you" in statement:
            print("I was developed by sudheer")
            speak("I was developed by sudheer")

        elif 'current time' in statement:
            strTime=datetime.datetime.now().strftime("%H:%M:%S")
            speak(f"the time is {strTime}")

```
